```python
def get_title(soup):
    try:
        title = soup.find("span",attrs={"class":'B_NuCI'}).get_text(strip=True)
    except AttributeError:
        title = ""
    return title
```

```python
def get_price(soup):
    try:
        price = soup.find("div",attrs={"class":'_30jeq3 _16Jk6d'}).get_text(strip=True)
        price = price[1:]
    except AttributeError:
        price = ""
    return price
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
URL = "https://www.flipkart.com/search?q=phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
```

```python
webpage = requests.get(url=URL, headers=HEADERS)
```

```python
soup = BeautifulSoup(webpage.content, 'html.parser')
```

```python
links = soup.find_all("a",attrs={'class':'_1fQZEK'})
```

```python
links_list=[]
for link in links:
    links_list.append(link.get("href"))
```

```python
dct = {"title":[],"price":[],"avg_rating":[],"total_ratings_and_reviews":[],"avg_camera_rating":[],"avg_battery_rating":[],"avg_display_rating":[],"avg_design_rating":[]}
```

```python
for link in links_list:
    new_webpage = requests.get("https://flipkart.com"+link,headers=HEADERS)
    new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
    
    dct["title"].append(get_title(new_soup))
    dct["price"].append(get_price(new_soup))
    dct["avg_rating"].append(get_avg_rating(new_soup))
    dct["total_ratings_and_reviews"].append(get_total_ratings_and_reviews(new_soup))
    dct["avg_camera_rating"].append(get_avg_rating_specific(new_soup,0))
    dct["avg_battery_rating"].append(get_avg_rating_specific(new_soup,1))
    dct["avg_display_rating"].append(get_avg_rating_specific(new_soup,2))
    dct["avg_design_rating"].append(get_avg_rating_specific(new_soup,3))
```

```python
dct
```

```python
flipkart_df = pd.DataFrame.from_dict(dct)
flipkart_df.sort_values(by='avg_rating',ascending=False)
flipkart_df.to_csv("flipkart_best_phones.csv",header=True,index = False)
```
